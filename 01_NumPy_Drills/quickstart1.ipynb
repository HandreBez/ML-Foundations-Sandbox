```python
import torch
import time

# Create two large random matrices (10,000 x 10,000)
size = 10000
a = torch.randn(size, size)
b = torch.randn(size, size)

# 1. Test on CPU
print(f"Running {size}x{size} matrix multiplication on CPU...")
start_time = time.time()
result_cpu = torch.matmul(a, b)
cpu_time = time.time() - start_time
print(f"CPU Time: {cpu_time:.4f} seconds")

# 2. Test on GPU (RTX 3070)
if torch.cuda.is_available():
    # Move matrices to GPU
    a_gpu = a.to('cuda')
    b_gpu = b.to('cuda')
    
    # Warm-up (to initialize the CUDA kernels)
    _ = torch.matmul(a_gpu, b_gpu)
    
    print(f"\nRunning {size}x{size} matrix multiplication on GPU...")
    start_time = time.time()
    result_gpu = torch.matmul(a_gpu, b_gpu)
    
    # Ensure all GPU operations are finished before stopping the clock
    torch.cuda.synchronize() 
    gpu_time = time.time() - start_time
    
    print(f"GPU Time: {gpu_time:.4f} seconds")
    print(f"Speedup: {cpu_time / gpu_time:.1f}x faster!")
else:
    print("\nGPU not detected. Check your drivers/installation.")
```
