```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y = np.array([[0],[1],[1],[0]], dtype=float)

# Network: 2 inputs → 2 hidden neurons → 1 output
np.random.seed(42)
W1 = np.random.randn(2, 2)   # (2 inputs, 2 hidden)
b1 = np.zeros((1, 2))
W2 = np.random.randn(2, 1)   # (2 hidden, 1 output)
b2 = np.zeros((1, 1))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Forward pass
z1 = X @ W1 + b1      # (4, 2)
a1 = sigmoid(z1)       # (4, 2)
z2 = a1 @ W2 + b2     # (4, 1)
a2 = sigmoid(z2)       # (4, 1) — this is your prediction



print(a2)
```

```python
loss = np.mean((a2 - y) ** 2)
print(loss)
```

```python
def sigmoid_deriv(z):
    return sigmoid(z) * (1 - sigmoid(z))

# Backprop
dL_da2 = 2 * (a2 - y) / 4          # derivative of loss w.r.t. output
dL_dz2 = dL_da2 * sigmoid_deriv(z2) # through the sigmoid
dL_dW2 = a1.T @ dL_dz2             # gradient for W2
dL_db2 = np.sum(dL_dz2, axis=0)    # gradient for b2

dL_da1 = dL_dz2 @ W2.T             # push error back to hidden layer
dL_dz1 = dL_da1 * sigmoid_deriv(z1)
dL_dW1 = X.T @ dL_dz1              # gradient for W1
dL_db1 = np.sum(dL_dz1, axis=0)    # gradient for b1
```

```python
lr = 1.0  # learning rate

W2 -= lr * dL_dW2
b2 -= lr * dL_db2
W1 -= lr * dL_dW1
b1 -= lr * dL_db1
```

```python
for epoch in range(100000):
    # forward
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)
    
    # loss
    loss = np.mean((a2 - y) ** 2)
    
    # backprop
    dL_da2 = 2 * (a2 - y) / 4
    dL_dz2 = dL_da2 * sigmoid_deriv(z2)
    dL_dW2 = a1.T @ dL_dz2
    dL_db2 = np.sum(dL_dz2, axis=0)
    dL_da1 = dL_dz2 @ W2.T
    dL_dz1 = dL_da1 * sigmoid_deriv(z1)
    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0)
    
    # update
    W2 -= lr * dL_dW2
    b2 -= lr * dL_db2
    W1 -= lr * dL_dW1
    b1 -= lr * dL_db1
    
    if epoch % 1000 == 0:
        print(f"epoch {epoch}  loss {loss:.4f}")
```
